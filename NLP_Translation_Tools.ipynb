```python
import pandas as pd
from googletrans import Translator
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/tale_of_kieu_vi.csv')
```

```python
translator = Translator()
```

```python
df['translated_en'] = df['sentence'].apply(lambda x: translator.translate(x, dest='en').text)
```

```python
# Save the translated DataFrame to a new CSV file
df.to_csv('/content/drive/MyDrive/data/tale_of_kieu_en_v2.csv', index=False)
```
